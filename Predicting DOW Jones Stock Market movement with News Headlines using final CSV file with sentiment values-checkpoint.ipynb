```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("combined with sentiment.csv",parse_dates=[0])
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(['Top1', 'Top2', 'Top3', 'Top4', 'Top5',
       'Top6', 'Top7', 'Top8', 'Top9', 'Top10', 'Top11', 'Top12', 'Top13', 'Top14', 'Top15', 'Top16', 'Top17', 'Top18',
            'Top19', 'Top20', 'Top21', 'Top22', 'Top23', 'Top24', 'Top25'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
#extracting date, label, and sentiment scores
first = data.loc[ : , ['Date', 'Label'] ]
first
cols = data.filter(regex=("CompoundValue"))
cols.head()
sentiment_data = first.merge(cols, how='inner', left_index = True, right_index=True)
sentiment_data.head()
```

```python
# Push the Label column to the end of the dataframe
cols = list(sentiment_data)
cols.append(cols.pop(cols.index('Label')))
sentiment_data = sentiment_data.loc[:, cols]
sentiment_data.head()
```

```python
stock_data = pd.read_csv('DJIA_table_with_start.csv', parse_dates=[0])
stock_data.head()
```

```python
# Create a dataframe by merging the sentiment scores of healines and the stock prices dataframes
merged_df = pd.merge(stock_data, sentiment_data, on='Date', how='inner')

# Check the shape of the merged dataframe
print(merged_df.shape)
```

```python
merged_df.head()
```

```python
corr_matrix = merged_df.corr()
```

```python
corr_matrix['Label'].sort_values(ascending = False)
```

start_trend =  (Closing Value of Previous day stock market - Opening Value of current day stock market)
'start_trend' value is a binary variable with values:
1 = positive trend 
0 = negative trend

The corelation matrix shows that the start_trend value is the most important predictor of the target variable followed by the sentiment value(compound value) of the Top19, Top13, Top3, Top15, Top9, Top8,Top1, Top2, Top6 headlines and Close, Adj Close values.

```python
merged_df.dtypes
```

```python
# Change the datatype of the volume column to float

merged_df['Volume'] = merged_df['Volume'].astype(float)
```

```python
# Creating train and test sets with data from dates 2008-08-08 to 2014-12-31 as Training Set and Test Set is
# the following two years data (from 2015-01-02 to 2016-07-01). This is roughly a 80%/20% split.
train_set = merged_df[merged_df['Date'] < '2015-01-01']
test_set = merged_df[merged_df['Date'] > '2014-12-31']
```

```python
X_train, y_train = train_set.iloc[:,:-1],train_set.iloc[:,-1]
X_test, y_test = test_set.iloc[:,:-1],test_set.iloc[:,-1]
```

```python
X_train=X_train.drop(['Date'],axis=1)
X_test=X_test.drop(['Date'],axis=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state = 42)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train, cv=3, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_train_pred)
```

```python
y_train_perfect_predictions = y_train
confusion_matrix(y_train, y_train_perfect_predictions)
```

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y_train, y_train_pred)
```

```python
recall_score(y_train, y_train_pred)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(random_state=42)
y_train_pred_rf = cross_val_predict(forest_clf, X_train, y_train, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_train_pred_rf)
```

```python
from sklearn.metrics import precision_score
precision_score(y_train, y_train_pred_rf)
```
